```python
# Calculate the percentage of missing values for each column
missing_percentage = (df.isnull().sum() / len(df)) * 100

# Filter columns with missing percentage > 30%
columns_to_drop = missing_percentage[missing_percentage > 30].index

# Drop those columns from the DataFrame
df_cleaned = df.drop(columns=columns_to_drop)

# Optional: Check the shape of cleaned DataFrame
print(f"Original shape: {df.shape}")
print(f"Cleaned shape: {df_cleaned.shape}")

```

```python
import matplotlib.pyplot as plt

# Sort values for better visualization (optional)
missing_percentage_sorted = missing_percentage.sort_values(ascending=False)

missing_percentage_sorted = missing_percentage_sorted[missing_percentage_sorted > 0]

# Plotting
plt.figure(figsize=(14, 6))
missing_percentage_sorted.plot(kind='bar', color='skyblue')

plt.title('Missing Value Percentage per Column')
plt.ylabel('Percentage (%)')
plt.xlabel('Columns')
plt.xticks(rotation=90)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
# Only numeric columns are useful for outlier detection
numeric_df = df_cleaned.select_dtypes(include=['number'])
```

```python
# IQR method to detect outliers
def detect_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (data < lower_bound) | (data > upper_bound)

# Detect outliers for all numeric columns
outliers = detect_outliers_iqr(numeric_df)

# Count how many outliers per column
outlier_counts = outliers.sum().sort_values(ascending=False)

print(outlier_counts[outlier_counts > 0])  # Show columns with outliers
```

```python
import matplotlib.pyplot as plt

# Choose two columns to plot (adjust based on your data)
x_col = outlier_counts.index[0]  # most outlier-prone feature
y_col = 'TARGET' if 'TARGET' in numeric_df.columns else numeric_df.columns[1]

plt.figure(figsize=(10, 6))
plt.scatter(numeric_df[x_col], numeric_df[y_col], 
            c=outliers[x_col].map({True: 'red', False: 'blue'}), alpha=0.5)

plt.xlabel(x_col)
plt.ylabel(y_col)
plt.title(f'Scatter Plot with Outliers Highlighted ({x_col} vs {y_col})')
plt.grid(True)
plt.show()
```

```python

# Segmenting
df_default = df_cleaned[df_cleaned['TARGET'] == 1]   # Payment difficulties
df_non_default = df_cleaned[df_cleaned['TARGET'] == 0]  # All others
```

```python
# Numeric columns only
numeric_cols = df_cleaned.select_dtypes(include='number').columns.drop('TARGET', errors='ignore')

# Correlation with TARGET on full dataset
cor_full = df_cleaned[numeric_cols].corrwith(df_cleaned['TARGET']).sort_values(key=abs, ascending=False)

# Correlation within defaulters (if target is constant, correlation won't work)
if df_default['TARGET'].nunique() > 1:
    cor_default = df_default[numeric_cols].corrwith(df_default['TARGET']).sort_values(key=abs, ascending=False)
else:
    cor_default = pd.Series(dtype=float)

# Correlation within non-defaulters (again, needs variance)
if df_non_default['TARGET'].nunique() > 1:
    cor_non_default = df_non_default[numeric_cols].corrwith(df_non_default['TARGET']).sort_values(key=abs, ascending=False)
else:
    cor_non_default = pd.Series(dtype=float)
```
